```python
numerical_features=[]
categorical_features=[]
for x in df.columns:
  if df[x].dtype==object and x!="Approved_Flag":
    categorical_features.append(x)
  elif df[x].dtype!=object and x not in ["PROSPECTID","Approved_Flag"]:
    numerical_features.append(x)
```

```python
print(categorical_features)
print(numerical_features)
```

```python
from scipy.stats import chi2_contingency
#Statement:finding association between features and target variable
#H0:Feature and Target Variable is not Associated
#H1:Feature and Target Variable is Associated
#Significance Value:0.05(prob of Type 1 Error)
#p-value(prob of observed data when H0 is true)
cat_features_tobe_kept=[]
cat_features_tobe_removed=[]
for col in categorical_features:
  chi2_val,p_val,_,_=chi2_contingency(pd.crosstab(df[col],df["Approved_Flag"]))
  print(col,"---",p_val)
  if p_val<0.05:
    cat_features_tobe_kept.append(col)
  else:
    cat_features_tobe_removed.append(col)
```

```python
print(cat_features_tobe_kept)
print(cat_features_tobe_removed)
```

```python
corr_matrix=df[numerical_features].corr()
corr_matrix
```

```python
removed_multicoll_features=[]
i=1
features=list(corr_matrix.columns)
for x in features:
  for j in range(0,len(features)):
    if x!=features[j]:
        c=corr_matrix.loc[x,features[j]]
        if c>0.8 or c<-0.8:
                    removed_multicoll_features.append(features[j])
                    features.remove(features[j])
                    break


```

```python
from scipy.stats import f_oneway
```

```python
#H0:features is not associated with target variable
#H1:Feature is asscociated with target variable

#this is basic assumed H0,H1,but actual is
#statement:does feature values making changes in target variable
#H0:all groups same(target variable doesnt change on feature value)
#H1:all groups are not same(feature value shows impact on target variable)
```

```python
numerical_features_final=[]
num_features_removed_anova=[]
for col in numerical_features_aftermulcoll:
  p1_group=[val for val,group in zip(df[col],df["Approved_Flag"]) if group=="P1"]
  p2_group=[val for val,group in zip(df[col],df["Approved_Flag"]) if group=="P2"]
  p3_group=[val for val,group in zip(df[col],df["Approved_Flag"]) if group=="P3"]
  p4_group=[val for val,group in zip(df[col],df["Approved_Flag"]) if group=="P4"]

  f_stat,p_val=f_oneway(p1_group,p2_group,p3_group,p4_group)
  if p_val<0.05:
    numerical_features_final.append(col)
  else:
    num_features_removed_anova.append(col)
```

```python
#Building a straightforward pipeline for all data transformations
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer,TransformedTargetRegressor
from sklearn.preprocessing import FunctionTransformer
```

```python
def transform(df):
     df=df[features_tobe_kept]
     return df
```

```python
features_transformer=FunctionTransformer(transform)
```

```python
def education_encoder(X):
     education_mapping={
         "SSC":0,
         "12TH":1,
          "UNDER GRADUATE":2,
         "GRADUATE":2,
         "OTHERS":0,
         "POST-GRADUATE":3,
         "PROFESSIONAL":4
      }

     X["EDUCATION"]=X["EDUCATION"].map(education_mapping).astype("int32")
     return X
```

```python
education_transformer=FunctionTransformer(education_encoder)
```

```python
def get_dummies(df_temp):
  #features=['MARITALSTATUS', 'GENDER', 'last_prod_enq2', 'first_prod_enq2']
  return pd.get_dummies(df_temp,dtype=int)
```

```python
onehotencoder_dummies=FunctionTransformer(get_dummies)
```

```python
encoder=ColumnTransformer(transformers=[("education_encoder",education_transformer,["EDUCATION"]),('onehotencoder',onehotencoder_dummies,['MARITALSTATUS', 'GENDER', 'last_prod_enq2', 'first_prod_enq2']),("num_features","passthrough",numerical_features_final)])
```

```python
model=Pipeline([("transform",features_transformer),("encoder",encoder),("xgboost",xgb_rs_be)])
```
